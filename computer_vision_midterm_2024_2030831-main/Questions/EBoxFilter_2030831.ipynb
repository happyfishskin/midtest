<h1>Integral Image</h1><div class="question_body"><p> Given is the following grayscale image of size 5x7.</p><div style="text-align:center;display:block"><table " ><tbody><tr><td>122</td><td>142</td><td>242</td><td>57</td><td>203</td><td>232</td><td>30</td></tr><tr><td>141</td><td>5</td><td>79</td><td>106</td><td>138</td><td>65</td><td>230</td></tr><tr><td>34</td><td>149</td><td>203</td><td>129</td><td>141</td><td>34</td><td>240</td></tr><tr><td>214</td><td>220</td><td>243</td><td>150</td><td>250</td><td>140</td><td>246</td></tr><tr><td>252</td><td>9</td><td>154</td><td>68</td><td>239</td><td>26</td><td>10</td></tr></tbody></table></div><p>The integral image <emph>s</emph> is defined by the following equation:</p><code>s[x,y] = img[x,y] + s[x-1,y] + s[x,y-1] - s[x-1,y-1]</code></div><!-- end of question_body -->

<div class="question_marks"><span class="mark_num" style="text-align:right">10</span> marks</div><!-- end of question_marks -->

<p>Calculate the integral image for this input image and add it into the table below</p><p>Enter the maximum value in the integral image that is less than 326 into the inputbox.</p><p>Maximum value in the integral image that is less than 326 <input type="number"></input>

<p>Show your work and explain your answer in this cell. You can add additional markdown and code cells, if you want to after this cell.</p>

```python
import numpy as np

# 定義影像矩陣
img = np.array([
    [122, 142, 242, 57, 203, 232, 30],
    [141, 5, 79, 106, 138, 65, 230],
    [34, 149, 203, 129, 141, 34, 240],
    [214, 220, 243, 150, 250, 140, 246],
    [252, 9, 154, 68, 239, 26, 10]
])

# 初始化積分影像矩陣
s = np.zeros_like(img, dtype=int)

#計算積分影像
for x in range (img.shape[0]):
    for y in range (img.shape[1]):
        s[x, y] = img[x, y]

        if x > 0:
            s[x, y] += s[x-1, y]
        if y > 0:
            s[x, y] += s[x, y-1]
        if x > 0 and y > 0:
            s[x, y] -= s[x-1, y-1]

max_value_below_326 = s[s < 326].max()
print("Maximum value in the integral image that is less than 326:", max_value_below_326)
```

<h1>Horizontal Box Filter</h1><div class="question_body"><p>A horizontal box filter of size <code>8</code> is defined as follows:</p><table " ><tbody><tr><td>1</td><td>1</td><td>1</td><td>1</td><td>1</td><td>1</td><td>1</td><td>1</td></tr><tr><td>1</td><td>1</td><td>1</td><td>1</td><td>1</td><td>1</td><td>1</td><td>1</td></tr><tr><td>1</td><td>1</td><td>1</td><td>1</td><td>1</td><td>1</td><td>1</td><td>1</td></tr><tr><td>1</td><td>1</td><td>1</td><td>1</td><td>1</td><td>1</td><td>1</td><td>1</td></tr><tr><td>0</td><td>0</td><td>0</td><td>0</td><td>0</td><td>0</td><td>0</td><td>0</td></tr><tr><td>0</td><td>0</td><td>0</td><td>0</td><td>0</td><td>0</td><td>0</td><td>0</td></tr><tr><td>0</td><td>0</td><td>0</td><td>0</td><td>0</td><td>0</td><td>0</td><td>0</td></tr><tr><td>0</td><td>0</td><td>0</td><td>0</td><td>0</td><td>0</td><td>0</td><td>0</td></tr></tbody></table><p>The filter is applied to the pixel at offset x = 4, y = 4.</p><p>You can pad the image by adding rows and columns of 0 if needed to calculate the value of the boxfilter along the edges of the image.</p></div><!-- end of question_body -->

<p>Calculate the output of the horizontal box filter for the image described above.</p><p>Enter the maximum value in the box filter output that is less than 0 into the inputbox.</p><p>Maximum value in the integral image that is less than 0 <input type="number"></input>

<p>Show your work and explain your answer in this cell. You can add additional markdown and code cells, if you want to after this cell.</p>

```python
import numpy as np

# 定義影像（示例影像矩陣，5x7影像）
img = np.array([
    [122, 142, 242, 57, 203, 232, 30],
    [141, 5, 79, 106, 138, 65, 230],
    [34, 149, 203, 129, 141, 34, 240],
    [214, 220, 243, 150, 250, 140, 246],
    [252, 9, 154, 68, 239, 26, 10]
])

# 定義濾波器
filter_size = 8
box_filter = np.zeros((filter_size, filter_size))
box_filter[:4, :] = 1  # 前四行為 1，後四行為 0\


# 計算濾波器輸出
output = np.zeros_like(img, dtype=float)
img_padded = np.pad(img, ((4, 4), (4, 4)), 'constant')  # 使用0填充影像

# 對每個位置應用濾波器
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        region = img_padded[i:i+filter_size, j:j+filter_size]
        output[i, j] = np.sum(region * box_filter)

# 找出小於 0 的最大值
max_value_below_0 = output[output < 0].max() if np.any(output < 0) else None
print("積分影像中小於 0 的最大值:", max_value_below_0)


```
