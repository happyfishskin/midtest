```python

import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

# 假設img是已載入的影像
def my_sobel(img, sobel_1, sobel_2, ksize=3):
    """
    img 是原始影像。
    sobel_1 和 sobel_2 是 Sobel 運算子的 X 和 Y 方向參數。
    如果 X 或 Y 方向的梯度非零，則應用 Sobel 濾波器來計算影像在該方向上的梯度。
    最後，將兩個方向上的 Sobel 運算結果相乘，返回結果。
    """
    img2 = img.copy()
    x_order_1, y_order_1 = sobel_1
    if (x_order_1 > 0) or (y_order_1 > 0):
        img2 = np.abs(cv.Sobel(img, cv.CV_64F, x_order_1, y_order_1, ksize=3))

    img3 = img.copy()
    x_order_2, y_order_2 = sobel_2
    if (x_order_2 > 0) or (y_order_2 > 0):
        img3 = np.abs(cv.Sobel(img, cv.CV_64F, x_order_2, y_order_2, ksize=3))

    return img2 * img3

# 載入影像
img = cv.imread('s.jpg', cv.IMREAD_GRAYSCALE)

# 應用不同的參數集
sobel_results = []
sobel_results.append(my_sobel(img, (1,0), (0,0)))
sobel_results.append(my_sobel(img, (0,1), (0,0)))
sobel_results.append(my_sobel(img, (2,0), (0,0)))
sobel_results.append(my_sobel(img, (0,2), (0,0)))
sobel_results.append(my_sobel(img, (1,1), (0,0)))
sobel_results.append(my_sobel(img, (1,0), (0,1)))
sobel_results.append(my_sobel(img, (0,2), (2,0)))
sobel_results.append(my_sobel(img, (1,0), (2,0)))
#因為原圖和原圖相乘了
sobel_results.append(my_sobel(img, (0,0), (0,0)))

# 顯示結果
plt.figure(figsize=(12, 12))
for i in range(9):
    plt.subplot(3, 3, i + 1)
    plt.imshow(sobel_results[i], cmap='gray')
    plt.xticks([])  # 隱藏 x 軸刻度
    plt.yticks([])  # 隱藏 y 軸刻度
    plt.title(f'Parameter set {i+1}')
plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# 加載一個範例影像（可替換為你的影像）
image = cv.imread('s.jpg', cv.IMREAD_GRAYSCALE)

# 對X方向應用Sobel運算子
sobel_x = cv.Sobel(image, cv.CV_64F, 1, 0, ksize=3)

# 對Y方向應用Sobel運算子
sobel_y = cv.Sobel(image, cv.CV_64F, 0, 1, ksize=3)

# 將Sobel X和Y結果相乘
sobel_x_mult_y = sobel_x * sobel_y

# 顯示結果
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.title('Sobel X')
plt.imshow(sobel_x, cmap='gray')

plt.subplot(1, 3, 2)
plt.title('Sobel Y')
plt.imshow(sobel_y, cmap='gray')

plt.subplot(1, 3, 3)
plt.title('Sobel X * Y')
plt.imshow(sobel_x_mult_y, cmap='gray')

plt.show()
```
